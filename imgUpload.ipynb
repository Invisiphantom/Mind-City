```python
import os
import chardet
import requests

all_files = []
for root, dirs, files in os.walk("."):
    for file in files:
        if file.endswith(".md"):
            all_files.append(os.path.join(root, file))

# 从all_files中移除非UTF-8编码的文件
for file in all_files:
    with open(file, 'rb') as f:
        content = f.read()
        encoding = chardet.detect(content)['encoding']
        if encoding != 'utf-8':
            all_files.remove(file)
            print(f"remove {file} because of encoding {encoding}")
```

```python
# ![image.png](https://s2.loli.net/2024/03/21/p7n9HotDyWjd1ri.png =550x) 
# 提取其中的图片链接, 并下载到本地
# 文件名为 2024_03_21_p7n9HotDyWjd1ri.png


def parse_img_url(line):
    url = line.split("(")[1].split(")")[0]
    if "=" in url:
        url = url.split("=")[0].strip()
    return url

def parse_img_name(url):
    return "_".join(url.split("/")[-4:])


def download_image(url, name):
    r = requests.get(url)
    with open(name, "wb") as f:
        f.write(r.content)

def pack_img_path(name):
    return os.path.join(os.getcwd(), name)


def upload_images(img_path_list):
    PicGo_url = "http://127.0.0.1:36677/upload"
    data = {"list": img_path_list}
    r = requests.post(PicGo_url, json=data)
    return r.json()["result"][0]



def get_my_url(line):
    url = parse_img_url(line)
    name = parse_img_name(url)
    download_image(url, name)
    path = pack_img_path(name)
    img_path_list=[path]
    res_url = upload_images(img_path_list)
    return url, res_url
    
```

```python

for file in all_files:
    
    with open(file, "r", encoding="utf-8") as f:
        lines = f.readlines()
    with open(file, "w", encoding="utf-8") as f:
        for line in lines:
            if "s2.loli.net" in line:
                print(f"Start: {file}  {line}")
                old_url, new_url = get_my_url(line)
                line = line.replace(old_url, new_url)
                print(f"Finish: {file}  {old_url} -> {new_url}")
            f.write(line)

```

```python
# 清除当前目录下的图片

for file in os.listdir("."):
    if file.endswith(".png") or file.endswith(".jpg"):
        os.remove(file)
```
